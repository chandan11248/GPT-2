```python
import torch
import torch.nn as nn

# Define a simple model
class CustomModel(nn.Module):
    def __init__(self, input_dim, output_dim):
        super().__init__()
        self.linear = nn.Linear(input_dim, output_dim)
        self.activation = nn.ReLU()

    def forward(self, x):
        x = self.linear(x)  # Linear transformation
        x = self.activation(x)  # Apply ReLU activation
        return x

# Instantiate the model
model = CustomModel(3, 2)

# Input tensor
x = torch.tensor([[1.0, 2.0, 3.0], [1.9, 2.6, 39.0]])

# Forward pass
output = model(x)  # This calls model.__call__(x), which invokes forward(x)
print(output)
```
